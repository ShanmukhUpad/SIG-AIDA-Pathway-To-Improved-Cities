```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../annual_boardings.csv")
for col in df:
    df[col] = df[col].astype(str).str.replace(',', '', regex=False).astype(int)
    
plt.plot(df['year'], df['bus'], linestyle='-', color="red")
plt.plot(df['year'], df['paratransit'], linestyle='-', color="yellow")
plt.plot(df['year'], df['rail'], linestyle='-', color="green")
plt.plot(df['year'], df['total'], linestyle='-', color="blue")

plt.xlabel("Year")
plt.ylabel("Boardings")
plt.show()
```
