```python
df['Date'] = pd.to_datetime(df['Date'])

df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year


df = df[['Date', 'Month', 'Year', 'Community Area', 'Primary Type', 'Arrest', 'Domestic']]


df = df[df['Community Area'].notna()]


agg = df.groupby(['Community Area', 'Year', 'Month', 'Primary Type']).size().reset_index(name='Count')


agg.to_csv("crime_monthly.csv", index=False)
```

```python
pivot = agg.pivot_table(
    index=['Community Area', 'Year', 'Month'],
    columns='Primary Type',
    values='Count',
    fill_value=0
).reset_index()

pivot.columns.name = None
pivot.to_csv("crime_monthly_pivot.csv", index=False)
```

```python
crime_cols = [c for c in pivot.columns if c not in ['Community Area', 'Year', 'Month']]

for col in crime_cols:
    pivot[f'{col}_lag1'] = pivot.groupby('Community Area')[col].shift(1)
    pivot[f'{col}_lag3'] = pivot.groupby('Community Area')[col].shift(3)

pivot = pivot.dropna()
```
